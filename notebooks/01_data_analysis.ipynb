```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("../data/zameen_rentals_data.csv")
df.head()

# Info and missing values
print(df.info())
print(df.isnull().sum())

# Describe numerical columns
print(df.describe())

# Visualizations
sns.histplot(df['rent'], bins=50)
plt.show()

sns.boxplot(x='city', y='rent', data=df)
plt.show()
```
